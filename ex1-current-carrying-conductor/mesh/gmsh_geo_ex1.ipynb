```python
import gmsh

gmsh.initialize()

gmsh.model.add("wire")

# -----------------------------
# Parameters
# -----------------------------

L = 1.0
cx = 0.5
cy = 0.5
r = 0.05

lc_air = 0.05
lc_wire = 0.01

# -----------------------------
# Outer square
# -----------------------------

square = gmsh.model.occ.addRectangle(
    0, 0, 0,
    L, L
)

# -----------------------------
# Wire
# -----------------------------

wire = gmsh.model.occ.addDisk(
    cx,
    cy,
    0,
    r,
    r
)

# -----------------------------
# Cut hole in square
# -----------------------------

air, _ = gmsh.model.occ.cut(
    [(2, square)],
    [(2, wire)],
    removeObject=True,
    removeTool=False
)

gmsh.model.occ.synchronize()

# -----------------------------
# Physical groups
# -----------------------------

gmsh.model.addPhysicalGroup(2, [air[0][1]], name="air")
gmsh.model.addPhysicalGroup(2, [wire], name="wire")

# -----------------------------
# Mesh sizes
# -----------------------------

gmsh.model.mesh.setSize(
    gmsh.model.getEntities(0),
    lc_air
)

# Make wire finer
wire_boundary = gmsh.model.getBoundary([(2, wire)], recursive=True)

gmsh.model.mesh.setSize(
    wire_boundary,
    lc_wire
)

# -----------------------------
# Generate
# -----------------------------

gmsh.model.mesh.generate(2)

gmsh.write("wire.msh")

gmsh.fltk.run()

gmsh.finalize()
```
